이로 인해 동물의 생명 손실뿐만 아니라, 교통사고를 유발하여 운전자의 안전을 위협하고, 생태계의 균형을 해치는 등 다양한 2차 피해가 발생합니다

2. 교통 특성과 로드킬 발생

- 반대로, 특정 조건에서는 속도가 너무 낮거나 차량 흐름이 적을 때 로드킬이 감소할 수도 있다.


3. 도로 지형과 로드킬 발생
- 도로의 기울기가 클수록 산악 지형과 같은 특성이 강해지며, 이러한 지역에서는 야생동물 이동 패턴에 영향을 미쳐 로드킬 발생 가능성이 달라질 수 있다.

### 가설 1번

### 결론 및 인사이트
1. 서식지(최단거리)와 발생건수는 약한 양의 상관관계를 보임. 이는 서식지가 도로와 가까울 수록 로드킬 발생건수가 증가하는 경향이 있음을 의미하지만, 상관계수가 낮으므로 주요원인이라는 단정지을 수 없다. 다른 요인도 고려할 필요가 있다.
2. 생태통로(최단거리)와 발생건수는 음의 상관관계를 보임. 생태통로는 야생동물이 도로를 횡단하지 않도록 유도하는 역활을 할 수 있다. 따라서 생태통로가 주변에 있으면 로드킬 발생을 줄이는 데 긍정적 영향이 미칠 수 있다. 
3. 하천(최단거리)와 발생건수는 약한 양의 상관관계를 보임. 하천이 도로 근처에 있으면, 야생동물이 도로를 횡단하기보다 하천을 따라 이동할 가능성이 높아 로드킬 발생일 줄일 수 있다는 점을 시사.

- 변수들 간의 상관행렬 분석 (서식지(최단거리), 생태통로(최단거리), 서식지(최단거리))\

1. 서식지(최단거리)와 하천(최단거리) 간의 상관계수는 -0.35로, 음의 상관관계를 보인다. 부분적인 반비례 관계가 있다는 것을 확인되므로 서식지와 하천이 분리되어 있을 가능성이 존재한다. 이는 동물들의 행동반경이 넓어지는 경향이 있다 라는 것을 의미하고, 그로 인해 도로를 횡단할 확률이 높아진다. 결론적으로 로드킬의 발생 확률을 높이는 요인으로 볼 수 있다

2. 하천(최단거리)와 생태통로(최단거리) 간의 상관계수는 0.16으로, 이는 두 변수 간에 약한 양의 상관관계를 나타낸다. 천과 생태통로는 어느 정도 연관이 있지만, 그 관계가 강하지 않다는 것을 의미. 하천와 생태통로가 가까운 위치에 있을 때, 동물들이 이동할 때 하천을 따라 생태통로를 활용할 가능성이 증가한다. 생태통로의 본연의 역할을 할 가능성이 높아지며, 야생동물들의 도로 횡단을 예방하고, 로드킬 위험을 줄 일 수 있는 요소로 작용할 수 있다.

- 반대로, 특정 조건에서는 속도가 너무 낮거나 차량 흐름이 적을 때 로드킬이 감소할 수도 있다.

### 결론 및 인사이트
- 반경5km내발생건수, 평균 총차량수, 평균 속도 

- 평균 총차량수와 평균 속도 간에는 음의 상관관계가 존재한다. 즉, 차량 수가 많을수록 평균 속도는 낮아지는 경향이 있다. 이는 교차로, 신호등 또는 도로의 혼잡도와 같은 요소들이 차량의 속도를 감소시키기 때문일 수 있다.
차량이 많으면 정체가 발생하고, 이에 따라 평균 속도가 감소하는 현상이 나타날 수 있다.

- 로드킬 발생 건수는 차량 통행량이 많은 도로에서 증가하는 경향을 보이며, 평균 속도가 낮을수록 로드킬 발생 확률이 높아질 수 있습니다. 이는 차량 통행량이 많을수록 동물이 도로를 횡단할 기회가 많아지고, 
낮은 속도에서는 동물이 도로를 횡단할 시간적 여유가 있을 수 있기 때문이다. 교차로, 도로의 혼잡도, 신호등과 같은 요소들이 속도에 영향을 미치면서, 로드킬 발생 건수와 관련된 중요한 변수를 형성할 수 있습니다.

- 해당 지점을 기준으로 주변 생태통로, 하천, 서식지의 위치 데이터를 확보합니다.

2. 거리 계산
- 로드킬 발생 지점과 주변 생태통로, 하천, 서식지 간의 최단거리를 계산 

3. 주변 환경 분석
- 로드킬 발생 지점을 중심으로 일정 반경 내에 하천, 생태통로, 서식지의 개수를 파악합니다.
- 동일 반경 내에서 로드킬 발생 건수를 집계합니다. 

```python
# 하천, 생태통로, 서식지 데이터 불러오기 
habitat = pd.read_csv("서식지.csv")
corridor = pd.read_csv('생태통로.csv')
river = pd.read_csv('하천.csv')
```

```python
river.head()
```

```python
from math import radians, sin, cos, sqrt, atan2

# Haversine 공식을 사용하여 두 지점 간의 거리를 계산하는 함수
def haversine_vectorized(lat1, lon1, lat2, lon2):
    R = 6371  # 지구 반지름 (km)
    lat1, lat2 = np.radians(lat1), np.radians(lat2)
    lon1, lon2 = np.radians(lon1), np.radians(lon2)
    dlat = lat2[None, :] - lat1[:, None]
    dlon = lon2[None, :] - lon1[:, None]
    a = np.sin(dlat / 2) ** 2 + np.cos(lat1[:, None]) * np.cos(lat2[None, :]) * np.sin(dlon / 2) ** 2
    c = 2 * np.arctan2(np.sqrt(a), np.sqrt(1 - a))
    return R * c

# 가장 가까운 지점까지의 최단 거리를 계산하는 함수
def calculate_shortest_distance(lat, lon, df):
    distances = haversine_vectorized(np.array([lat]), np.array([lon]), df['위도'].values, df['경도'].values)
    return np.min(distances)

# 하천의 경우 시작점과 끝점까지의 최단 거리를 계산하는 함수
def calculate_river_distance(lat, lon, df_river):
    start_distances = haversine_vectorized(np.array([lat]), np.array([lon]), df_river['시점위치(위도)'].values, df_river['시점위치(경도)'].values)
    end_distances = haversine_vectorized(np.array([lat]), np.array([lon]), df_river['종점위치(위도)'].values, df_river['종점위치(경도)'].values)
    return np.minimum(np.min(start_distances), np.min(end_distances))

# 각 행에 대해 최단 거리 계산 함수 정의
def find_nearest_distances(row, habitat_df, corridor_df, river_df):
    lat, lon = row['위도'], row['경도']
    
    # 서식지(최단거리)
    min_habitat_distance = calculate_shortest_distance(lat, lon, habitat_df)
    
    # 생태통로(최단거리)
    min_corridor_distance = calculate_shortest_distance(lat, lon, corridor_df)
    
    # 하천(최단거리)
    min_river_distance = calculate_river_distance(lat, lon, river_df)
    
    return pd.Series({
        '서식지(최단거리)': min_habitat_distance,
        '생태통로(최단거리)': min_corridor_distance,
        '하천(최단거리)': min_river_distance
    })

# total_report 데이터프레임에 최단 거리 칼럼 추가
total_report[['서식지(최단거리)', '생태통로(최단거리)', '하천(최단거리)']] = total_report.apply(
    find_nearest_distances, axis=1, habitat_df=habitat, corridor_df=corridor, river_df=river
)

# 반경 5km 내의 발생 건수를 계산하는 함수
def count_within_radius(lat, lon, df, radius=5.0):
    distances = haversine_vectorized(np.array([lat]), np.array([lon]), df['위도'].values, df['경도'].values)
    return np.sum(distances <= radius)

# 각 로드킬 발생 위치에 대해 반경 5km 내의 발생 건수 계산
total_report['반경5km내발생건수'] = total_report.apply(
    lambda row: count_within_radius(row['위도'], row['경도'], total_report), axis=1
)

# 반경 5km 내의 특정 지점 개수를 계산하는 함수
def count_within_radius(lat, lon, df, radius=5.0):
    distances = haversine_vectorized(np.array([lat]), np.array([lon]), df['위도'].values, df['경도'].values)
    return np.sum(distances <= radius)

# 각 로드킬 발생 위치에 대해 반경 5km 내의 서식지, 생태통로, 하천 개수 계산
def add_nearby_feature_counts(total_report, habitat, corridor, river):
    total_report['서식지 개수(5km)'] = total_report.apply(
        lambda row: count_within_radius(row['위도'], row['경도'], habitat), axis=1
    )
    total_report['생태통로 개수(5km)'] = total_report.apply(
        lambda row: count_within_radius(row['위도'], row['경도'], corridor), axis=1
    )

    return total_report

# Haversine 공식을 사용하여 두 지점 간의 거리를 계산하는 함수
def haversine(lat1, lon1, lat2, lon2):
    R = 6371  # 지구 반지름 (km)
    
    # 위도, 경도를 라디안으로 변환
    lat1, lon1, lat2, lon2 = map(radians, [lat1, lon1, lat2, lon2])
    
    # 차이 계산
    dlat = lat2 - lat1
    dlon = lon2 - lon1
    
    # Haversine 공식
    a = sin(dlat / 2)**2 + cos(lat1) * cos(lat2) * sin(dlon / 2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))
    
    # 거리 계산
    distance = R * c
    return distance

# 하천의 시작점과 끝점까지의 거리가 반경 5km 이내인 하천의 개수를 계산하는 함수
def count_rivers_within_radius(lat, lon, river_df, radius=5.0):
    count = 0
    for _, river in river_df.iterrows():
        # 하천의 시작점과 끝점 좌표
        start_lat, start_lon = river['시점위치(위도)'], river['시점위치(경도)']
        end_lat, end_lon = river['종점위치(위도)'], river['종점위치(경도)']
        
        # 시작점과 끝점까지의 거리 계산
        start_distance = haversine(lat, lon, start_lat, start_lon)
        end_distance = haversine(lat, lon, end_lat, end_lon)
        
        # 두 거리 중 가장 짧은 거리 계산
        min_distance = min(start_distance, end_distance)
        
        # 반경 내에 하천이 있을 경우 카운트 증가
        if min_distance <= radius:
            count += 1
    return count

# 각 로드킬 발생 위치에 대해 반경 5km 내의 하천 개수 계산
total_report['하천 개수(5km)'] = total_report.apply(
    lambda row: count_rivers_within_radius(row['위도'], row['경도'], river), axis=1
)

# 함수 실행
total_report = add_nearby_feature_counts(total_report, habitat, corridor, river)
```

```python
# 가장 가까운 도로의 정보를 찾는 함수
def find_nearest_road(lat, lon, df_road):
    distances = haversine_vectorized(np.array([lat]), np.array([lon]), df_road['위도'].values, df_road['경도'].values)
    nearest_idx = np.argmin(distances)
    return (
        df_road.iloc[nearest_idx]['평균 총차량수'],
        df_road.iloc[nearest_idx]['평균 속도'],
        df_road.iloc[nearest_idx]['기울기']
    )

# 각 로드킬 발생 위치에 대해 가장 가까운 도로의 정보 추가
def add_nearest_road_info(total_report, road):
    total_report[['평균 총차량수', '평균 속도', '기울기']] = total_report.apply(
        lambda row: pd.Series(find_nearest_road(row['위도'], row['경도'], road)),
        axis=1
    )
    return total_report

# 함수 실행
total_report = add_nearest_road_info(total_report, road)
```

### 결론 및 인사이트
1. 서식지(최단거리)와 발생건수는 약한 양의 상관관계를 보임. 이는 서식지가 도로와 가까울 수록 로드킬 발생건수가 증가하는 경향이 있음을 의미하지만, 상관계수가 낮으므로 주요원인이라는 단정지을 수 없다. 다른 요인도 고려할 필요가 있다.
2. 생태통로(최단거리)와 발생건수는 음의 상관관계를 보임. 생태통로는 야생동물이 도로를 횡단하지 않도록 유도하는 역활을 할 수 있다. 따라서 생태통로가 주변에 있으면 로드킬 발생을 줄이는 데 긍정적 영향이 미칠 수 있다. 
3. 하천(최단거리)와 발생건수는 약한 양의 상관관계를 보임. 하천이 도로 근처에 있으면, 야생동물이 도로를 횡단하기보다 하천을 따라 이동할 가능성이 높아 로드킬 발생일 줄일 수 있다는 점을 시사.

- 변수들 간의 상관행렬 분석 (서식지(최단거리), 생태통로(최단거리), 서식지(최단거리))\

1. 서식지(최단거리)와 하천(최단거리) 간의 상관계수는 -0.35로, 음의 상관관계를 보인다. 부분적인 반비례 관계가 있다는 것을 확인되므로 서식지와 하천이 분리되어 있을 가능성이 존재한다. 이는 동물들의 행동반경이 넓어지는 경향이 있다 라는 것을 의미하고, 그로 인해 도로를 횡단할 확률이 높아진다. 결론적으로 로드킬의 발생 확률을 높이는 요인으로 볼 수 있다

2. 하천(최단거리)와 생태통로(최단거리) 간의 상관계수는 0.16로, 아주 약한 음의 상관관계를 보인다. 변수 간 약한 반비례를 관계를 있다는 것을 의미한다. 즉, 두 요소가 서로 다르게 배치된 경우가 있음을 시사한다.하천과 생태통로가 가까이 위치하지 않다라는 것을 의미한다. 이는 만약 야생동물들이 하천을 따라 이동할 때 생태통로가 적거나 없는 경우에 다른 경로를 따라 이동할 가능성이 커지고, 도로를 횡단할 확률이 증가할 수 있다.
